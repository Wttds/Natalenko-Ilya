```python
from copy import deepcopy

# Функция: Печать шахматной доски |||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||


def print_the_board(massive):
    for line in massive:
        print(*line)


# Функция: Cчитает 0 и определяет их координаты |||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||


def return_free_cells(field: list, size: int):
    fc_arr = []
    for row in range(size):
        for col in range(size):
            if field[row][col] == "0":
                fc_arr.append((row, col))
    return fc_arr


# Функция: Пометка звёздочкой клеток под боем для фигуры Конь-Дракон (Вызывается при каждой поставленной фигуре) ||||||


def mark_dangerous_cells(field: list, size: int, x: int, y: int):
    field[x][y] = "#"
    for row in range(-1, 2):    # [Клетки вокруг фигуры]
        for col in range(-1, 2):
            if 0 <= x + row < size and 0 <= y + col < size and field[x + row][y + col] == "0":
                field[x + row][y + col] = "*"
    step = 2                    # [Клетки по диагонали]
    while step < size:
        for row in range(x - step, x + step + 1, step * 2):
            for col in range(y - step, y + step + 1, step * 2):
                if 0 <= row < size and 0 <= col < size and field[row][col] == "0":
                    field[row][col] = "*"
        step += 1

# Функция: проверяет правильность решения и возвращает координаты поставленных фигур, если решение верно. В противном |
# случае возвращает пустой список.                                                                                    |


def is_salvation(field, bin_str, free_cells):
    temp_board, salvation, out_coords = deepcopy(field), True, []
    for digit in range(len(bin_str)):
        if bin_str[digit] == "1":
            if temp_board[free_cells[digit][0]][free_cells[digit][1]] != "*":
                temp_x, temp_y = free_cells[digit][0], free_cells[digit][1]
                out_coords.append((temp_x, temp_y))
                mark_dangerous_cells(temp_board, len(temp_board), temp_x, temp_y)
            else:
                salvation = False
                break
    if salvation:
        return out_coords
    return []


def main():
    # Работа с данными ||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||

    f = open("input.txt")                       # Чтение и создание переменных по условию задачи
    temp_data = f.readline().rstrip().split()
    n, l, k = int(temp_data[0]), int(temp_data[1]), int(temp_data[2])
    board = [["0"] * n for _ in range(n)]       # Создание доски и заполнение её фигурами
    figures = []
    for i in range(k):
        a, b = list(map(int, f.readline().rstrip().split()))
        figures.append((a, b))
        mark_dangerous_cells(board, n, a, b)
    f.close()

    # Варианты ходов в виде бинарных строк. 1 занято, 0 - пусто (В каждом варианте только l) ||||||||||||||||||||||||||

    b_printed = False
    out = open("output.txt", "w")
    fc_array = return_free_cells(board, n)
    fc_number = len(fc_array)
    for num in range(1, 2 ** fc_number):
        if format(num, "b").count("1") == l:
            coords = figures + is_salvation(board, format(num, "b").rjust(fc_number, "0"), fc_array)
            if coords != figures:
                out.write(str(coords)[1:-1] + "\n")
                if not b_printed:      # Код на этой строке и ниже, отвечает только за вывод одного варианта на консоль
                    board_clone = deepcopy(board)
                    for pair in coords:
                        mark_dangerous_cells(board_clone, n, pair[0], pair[1])
                    print_the_board(board_clone)
                    b_printed = True


if __name__ == "__main__":
    main()
```
