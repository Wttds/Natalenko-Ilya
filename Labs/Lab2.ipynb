```python
from copy import deepcopy


def main():
    # Функция: Печать шахматной доски |||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||

    def print_the_board(massive):
        for line in massive:
            print(*line)

    # Функция: Cчитает 0 и определяет их координаты |||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||

    def return_free_cells(field: list, size: int):
        fc_arr = []
        for row in range(size):
            for col in range(size):
                if field[row][col] == "0":
                    fc_arr.append((row, col))
        return fc_arr

    # Функция: Пометка звёздочкой клеток под боем для фигуры Конь-Дракон (Вызывается при каждой поставленной фигуре) ||

    def mark_dangerous_cells(field: list, size: int, x: int, y: int):
        for row in range(-1, 2):    # [Клетки вокруг фигуры]
            for col in range(-1, 2):
                if 0 <= x + row < size and 0 <= y + col < size and field[x + row][y + col] == "0":
                    field[x + row][y + col] = "*"
        step = 2                    # [Клетки по диагонали]
        while step < size:
            for row in range(x - step, x + step + 1, step * 2):
                for col in range(y - step, y + step + 1, step * 2):
                    if 0 <= row < size and 0 <= col < size and field[row][col] == "0":
                        field[row][col] = "*"
            step += 1

    # Работа с данными ||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||

    f = open("input.txt")
    temp_data = f.readline().rstrip().split()
    n, l, k = int(temp_data[0]), int(temp_data[1]), int(temp_data[2])

    # Создание доски и заполнение её фигурами |||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||

    board = []
    for i in range(n):
        board.append(["0"] * n)
    for i in range(k):
        temp_data = f.readline().rstrip().split()
        a, b = int(temp_data[0]), int(temp_data[1])
        board[a][b] = "#"
        mark_dangerous_cells(board, n, int(temp_data[0]), int(temp_data[1]))

    # Варианты ходов в виде бинарных строк. 1 занято, 0 - пусто (В каждом варианте только l) ||||||||||||||||||||||||||

    board_printed = False
    out = open("output.txt", "w")
    fc_array = return_free_cells(board, n)
    fc_number = len(fc_array)
    nums = [str(format(num, 'b')).rjust(fc_number, '0') for num in range(3, 2 ** fc_number + 1)]
    nums = list(filter(lambda x: x.count('1') == l, nums))
    for binary in nums:

        # Проверка варианта на соответствие условиям ||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||

        temp_array, is_salvation, out_coords = deepcopy(board), True, []
        for digit in range(len(binary)):
            if binary[digit] == "1":
                if temp_array[fc_array[digit][0]][fc_array[digit][1]] != "*":
                    temp_x, temp_y = fc_array[digit][0], fc_array[digit][1]
                    temp_array[temp_x][temp_y] = "#"
                    out_coords.append((temp_x, temp_y))
                    mark_dangerous_cells(temp_array, n, temp_x, temp_y)
                else:
                    is_salvation = False
                    break
        if is_salvation:
            out.write(str(out_coords)[1:-1] + "\n")
            if not board_printed:
                print_the_board(temp_array)
                board_printed = True

if __name__ == "__main__":
    main()
```
