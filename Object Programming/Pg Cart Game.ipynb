```python
import pygame
import random

# Определение констант
WIDTH = 1200
HEIGHT = 800
FPS = 60

# Определение цветов
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Инициализация Pygame и создание окна
pygame.init()
pygame.mixer.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Catch Game")
clock = pygame.time.Clock()

# Класс для тележки, которую управляет игрок
class Cart(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load('Natalenko-Ilya/Object Programming/Pygame pics and sounds/телега.png').convert()
        self.image = pygame.transform.scale(self.image, (400, 160))  # изменение размера изображения телеги
        self.image.set_colorkey(WHITE)
        self.rect = self.image.get_rect()
        self.rect.centerx = WIDTH / 2
        self.rect.bottom = HEIGHT - 10
        self.speedx = 0

    def update(self):
        # Обновление скорости тележки
        self.speedx = 0
        keystate = pygame.key.get_pressed()
        if keystate[pygame.K_LEFT]:
            self.speedx = -8
        if keystate[pygame.K_RIGHT]:
            self.speedx = 8
        self.rect.x += self.speedx
        # Ограничение движения тележки по экрану
        if self.rect.right > WIDTH:
            self.rect.right = WIDTH
        if self.rect.left < 0:
            self.rect.left = 0


# Класс для падающих предметов
class Item(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.surf = pygame.Surface((10, 10))
        self.image = pygame.image.load('Natalenko-Ilya/Object Programming/Pygame pics and sounds/медведь.png')
        self.image = pygame.transform.scale(self.image, (50, 50))
        self.rect = self.image.get_rect()
        self.rect.x = random.randrange(WIDTH - self.rect.width)
        self.rect.y = random.randrange(-100, -40)
        self.speedy = random.randrange(1, 8)

    def update(self):
        # Обновление позиции падающего предмета на экране
        self.rect.y += self.speedy
        if self.rect.top > HEIGHT:
            self.rect.x = random.randrange(WIDTH - self.rect.width)
            self.rect.y = random.randrange(-100, -40)
            self.speedy = random.randrange(1, 8)


# Создание групп для всех спрайтов и для падающих предметов
all_sprites = pygame.sprite.Group()
items = pygame.sprite.Group()

# Создание тележки и добавление ее в все спрайты
cart = Cart()
all_sprites.add(cart)

# Основной игровой цикл
running = True
score = 0
while running:
    # Установление частоты кадров
    clock.tick(FPS)
    max_items = 10
    if len(items) == 0:
        for i in range(1 + score // 10):
            if i == max_items:
                break
            else:
                item = Item()
                items.add(item)
                all_sprites.add(item)
    # Обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Обновление всех спрайтов
    all_sprites.update()

    # Проверка столкновений тележки и падающих предметов
    hits = pygame.sprite.spritecollide(cart, items, True)
    for hit in hits:
        score += 10
    if score >= 100:
        print("Победа!")
        break

    # Отрисовка всех спрайтов
    screen.fill(BLACK)
    all_sprites.draw(screen)

    # Отображение очков на экране
    font = pygame.font.Font(None, 36)
    text = font.render("Score: " + str(score), True, WHITE)
    screen.blit(text, (10, 10))

    # Показ кадра
    pygame.display.flip()

# Завершение Pygame
pygame.quit()
```
