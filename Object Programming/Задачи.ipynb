```python
class Box:
    def __init__(self):
        self.box = set()

    def add(self, *elements):
        for elem in elements:
            self.box.add(elem)

    def empty(self):
        out = []
        for elem in self.box:
            out.append(elem)
        self.box.clear()
        return out

    def count(self):
        return len(self.box)


class Item:
    def __init__(self, name, value):
        self.name = name
        self.value = value


class ListBox(Box):
    def __init__(self):
        super().__init__()
        self.box = list(self.box)

    def add(self, *elements):
        for elem in elements:
            self.box.append(elem)

class DictBox(Box):
    def __init__(self):
        super().__init__()
        self.box = dict(self.box)

    def add(self, *elements):
        for elem in elements:
            self.box[elem] = elem


def repack_boxes(*boxes):
    summary = []
    for box in boxes:
        summary += box.empty()
    lengths, left = [len(summary) // len(boxes)] * len(boxes), len(summary) % len(boxes)
    while left > 0:
        for i in range(len(lengths)):
            lengths[i] += 1
            left -= 1
            if left == 0:
                break

    i = 0
    print(f"lengths={lengths}")
    for a in range(len(lengths)):
        for b in range(lengths[a]):
            boxes[a].add(summary[i])
            i += 1
```

```python
class User:
    def __init__(self, name, email, age):
        self.name = name
        self.email = email
        self.age = age

class NameNotUniqueException(Exception):
    pass

class AgeNotPositiveException(Exception):
    pass

class AgeUnder16Exception(Exception):
    pass

class InvalidEmailException(Exception):
    pass

def add_user_to_catalog(catalog, user):
    try:
        for u in catalog:
            if u.name == user.name:
                raise NameNotUniqueException("Name already exists in catalog")
            if user.age <= 0:
                raise AgeNotPositiveException("Age is not a positive integer")
            if user.age < 16:
                raise AgeUnder16Exception("User is under 16 years old")
            if '@' not in user.email or len(user.email.split('@')) != 2 or user.email.count('.') < 1:
                raise InvalidEmailException("Email address is invalid")
        catalog.append(user)
    except NameNotUniqueException as e:
        print("Error adding user to catalog:", e)
    except AgeNotPositiveException as e:
        print("Error adding user to catalog:", e)
    except AgeUnder16Exception as e:
        print("Error adding user to catalog:", e)
    except InvalidEmailException as e:
        print("Error adding user to catalog:", e)
```

## 3

Класс "Песня" (Song) - описывает свойства и методы для каждой отдельной песни, такие как название, длительность, исполнитель, альбом и другие метаданные.

Класс "Исполнитель" (Artist) - хранит информацию об исполнителях, включая их имена, биографии, жанры музыки, список песен и альбомов и т.д.

Класс "Альбом" (Album) - описывает альбомы с музыкой, включая название, дату выпуска, список песен, исполнителей и другие метаданные.

Класс "Плейлист" (Playlist) - содержит список песен, который пользователь создал и настроил для воспроизведения.

Связи между классами:

Песня связана с исполнителем, так как все песни кто-то исполняет. Один исполнитель может иметь много песен, но каждая песня может иметь 
толькоодного исполнителя. Таким образом, связь здесь однонаправленная. Альбом содержит список песен, и каждая песня связана с определенным альбомом. Альбом также связан с исполнителем, который записал этот альбом. Связи являются двунаправленными, т.к. мы можем получить список песен из альбома или же список альбомов, которые записал исполнитель. Плейлист связан с песнями, которые в нем находятся. Это однонаправленная связь, т.к. мы можем получать список песен из плейлиста, но не наоборот.

```python
class Song:
    def __init__(self, title, duration, artist, album):
        self.title = title
        self.duration = duration
        self.artist = artist
        self.album = album

class Artist:
    def __init__(self, name, bio, genre):
        self.name = name
        self.bio = bio
        self.genre = genre
        self.songs = []
        self.albums = []

class Album:
    def __init__(self, title, release_date, artist):
        self.title = title
        self.release_date = release_date
        self.artist = artist
        self.songs = []

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)
        
# Пример создания альбома и добавление его песен в плейлист
artist = Artist("Queen", "Legendary British rock band", "Rock")
album = Album("The Game", "June 1980", artist)

song1 = Song("Another One Bites the Dust", "3:35", artist, album)
song2 = Song("Crazy Little Thing Called Love", "2:42", artist, album)
song3 = Song("Save Me", "3:49", artist, album)

album.songs = [song1, song2, song3]
artist.albums.append(album)
artist.songs.extend(album.songs)

playlist1 = Playlist("My Awesome Playlist")
playlist1.add_song(song1)
playlist1.add_song(song2)
playlist1.add_song(song3)
```

```python
class Learner:
    def __init__(self):
        self.classes = []

    def enrol(self, course):
        self.classes.append(course)


class Teacher:
    def __init__(self):
        self.courses_taught = []


    def assign_teaching(self, course):
        self.courses_taught.append(course)


class Person:
    def __init__(self, name, surname, number, learner=None, teacher=None):
        self.name = name
        self.surname = surname
        self.number = number

        self.learner = learner
        self.teacher = teacher

    def enrol(self, course):
        if self.learner:
            self.learner.enrol(course)
        else:
            raise AttributeError("This person is not a learner")

    def assign_teaching(self, course):
        if self.teacher:
            self.teacher.assign_teaching(course)
        else:
            raise AttributeError("This person is not a teacher")

jane = Person("Jane", "Smith", "SMTJNX045", Learner(), Teacher())
jane.enrol('a_postgrad_course')
jane.assign_teaching('an_undergrad_course')
```
