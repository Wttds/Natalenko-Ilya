```python
class Box:
    def __init__(self):
        self.box = set()

    def add(self, *elements):
        for elem in elements:
            self.box.add(elem)

    def empty(self):
        out = []
        for elem in self.box:
            out.append(elem)
        self.box.clear()
        return out

    def count(self):
        return len(self.box)


class Item:
    def __init__(self, name, value):
        self.name = name
        self.value = value


class ListBox(Box):
    def __init__(self):
        super().__init__()
        self.box = list(self.box)

    def add(self, *elements):
        for elem in elements:
            self.box.append(elem)

class DictBox(Box):
    def __init__(self):
        super().__init__()
        self.box = dict(self.box)

    def add(self, *elements):
        for elem in elements:
            self.box[elem] = elem


def repack_boxes(*boxes):
    summary = []
    for box in boxes:
        summary += box.empty()
    lengths, left = [len(summary) // len(boxes)] * len(boxes), len(summary) % len(boxes)
    while left > 0:
        for i in range(len(lengths)):
            lengths[i] += 1
            left -= 1
            if left == 0:
                break

    i = 0
    print(f"lengths={lengths}")
    for a in range(len(lengths)):
        for b in range(lengths[a]):
            boxes[a].add(summary[i])
            i += 1
```
