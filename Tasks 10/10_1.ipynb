1. Первая производная и её график.

Первая: 6 * log(5 * x ** 3) * cos(6 * x) + 3 * sin(6 * x) / x

```python
import matplotlib.pyplot as plt
from sympy import diff, symbols, log, sin, cos
import numpy as np

x = symbols('x')
f_diff = diff(log(5 * x ** 3) * sin(6 * x))
print(f_diff)

x = np.linspace(0.1, 5, 1000)
y = np.array([6 * log(5 * a ** 3) * cos(6 * a) + 3 * sin(6 * a) / a for a in x])
plt.plot(x, y)
plt.show()
```

2. Наибольшее и наименьшее значения на отрезке.

```python
x = np.linspace(0.1, 5, 1000)
y = np.array([log(5 * a ** 3) * sin(6 * a) for a in x])

min_y = np.argmin(y)
min_x = x[min_y]
min_y = y[min_y]

max_y = np.argmax(y)
max_x = x[max_y]
max_y = y[max_y]

plt.plot(x, y)
plt.plot(min_x, min_y, 'o', color='blue')
plt.plot(max_x, max_y, 'o', color='red')
plt.show()
```

3. Касательная и нормаль к точке.

```python
x = np.linspace(0.1, 5, 1000)
y = np.array([log(5 * a ** 3) * sin(6 * a) for a in x])


def f(a):   # Функция
    return log(5 * a ** 3) * sin(6 * a)


x_0 = 1.45     # a) Случайная точка 1.31
y_0 = f(x_0)


def f_diff(a):  # b) Производная
    return 6 * log(5 * a ** 3) * cos(6 * a) + 3 * sin(6 * a) / a


y_diff_0 = f_diff(x_0)  # c) Значение производной в x_0

k = f(x_0) + f_diff(x_0) * (x - x_0)    # d) Касательная

n = y_0 - (1 / y_diff_0) * (x - x_0)   # Нормаль

plt.plot(x, y, c="black")   # Рисование функций
plt.plot(x, k, "b--")
plt.plot(x, n, "r-.")
plt.show()
```

4. Кассательное расслоение.

```python
x = np.linspace(0.1, 5, 1000)
y = np.array([log(5 * a ** 3) * sin(6 * a) for a in x])

def f(a):   # Функция
    return log(5 * a ** 3) * sin(6 * a)

def f_diff(a):  # Производная
    return 6 * log(5 * a ** 3) * cos(6 * a) + 3 * sin(6 * a) / a

plt.plot(x, y)

for x_0 in np.arange(0.3, 0.9, 0.1):
    y_0 = f(x_0)    # Значение функции в x_0
    y_diff_0 = f_diff(x_0)  # Значение производной в x_0
    k = f(x_0) + f_diff(x_0) * (x - x_0)    # Касательная
    plt.plot(x, k, "r--")

plt.show()
```
